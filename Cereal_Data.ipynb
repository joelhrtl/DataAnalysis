```python
# List of missing values by columns
df.isna().sum()
```

```python
# select the num columns
num_cols = df.select_dtypes(include=np.number).columns
num_cols
```

```python
# select cat columns
cat_cols = df.select_dtypes(include=object).columns
cat_cols
```

```python
# clean the num cols
for col in num_cols:
  df[col] = df[col].fillna(df[col].mean())
```

```python
# clean the cat cols
for col in cat_cols:
  df[col] = df[col].fillna(df[col].mode()[0])
```

```python
df.isna().sum()
```

```python
# Print first and last six rows
df.head(6)
df.tail(6)
```

```python
# Find the minimum calories cereal name
min_cer_name = df[df["calories"] == df["calories"].min()]
min_cer_name
```

```python
# Print the list of cereal name with sugar rate = 0
zero_sugars = df[df["sugars"] == 0][["name", "sugars"]]
zero_sugars
```

```python
# Print the list of cereal name with the following four conditions:
cer_4conditions = df[ (df["calories"]> df["calories"].mean())
                    & (df["protein"]> df["protein"].mean())
                    & (df["carbo"]< df["carbo"].mean())
                    & (df["potass"]< df["potass"].mean())]
cer_4conditions
```

```python
# Print the minimum of "Vitamins_rate & sodium_rate" by cereal type
Tab1 = df.groupby("type")[["vitamins", "sodium"]].min().reset_index()
Tab1
```

```python
# Print the minimum of "Vitamins_rate & sodium_rate" by cereal typ & mfr
Tab2 = df.groupby(["type", "mfr"])[["vitamins", "sodium"]].min().reset_index()
Tab2
```
